### 어떻게 시작해야 할까
1) 고객에게 어떤 경험을 제공할 수 있을까 <br>
> 결국 '사용자의 목적 달성에 도움을 줄 수 있는가' 를 고려해야 한다. 
* 유튜브의 경우 유튜브 시청자들을 위한 추천 서비스는 크리에이터와 유사한 크리에이터를 소개하거나, 기존 시청한 영상과 유사한 영상을 제안하는 것. <br>
<br>
* (수현's 경험담) 
    * 뉴스 사용자에게 현재 실시간으로 보도되고 있는 속보성 기사들, 현재 이슈가 되고있는 기사들을 볼 수 있게 제안하는 것이 될 수 있음.

2) 중장기적으로 비즈니스에 어떤 가치를 제공할 수 있는가<br>
> 개인화 서비스에 들어가는 비용이 적지 않으므로, 이를 통한 매출 증대나 충성 고객 창출과 같은 비즈니스적 효과가 있는지 검토해야 한다.
* (수현's 경험담)
    * 실제로 회사에서 뉴스를 사람이 직접 보고 기사를 선별해서 올리다가 자동화 시키는 데 6개월 남짓의 시간이 소요되었음. <br>
    여기에 투입된 개발자가 팀장급 2명, 팀원급 2명 정도 된다고하면 이들의 6개월치 연봉으로 환산하면 대~충 1억은 들어가지 않았을까 하는 지극히 개인적인 추정..

3) 고객이 기꺼이 제공할 수 있는 데이터인가<br>
> 인구통계학적 데이터(개인정보) 는 비교적 수집하기 쉬운 데이터에 속함. 하지만 실제 사용자의 감정을 파악하기 위해서는 콘텐츠 선호도 데이터를 수집해야 하는데, 이는 서비스 제공자와 사용자 모두에게 많은 노력이 필요함. <br>
따라서, 행동을 요구하는 데이터의 경우 고객이 기꺼이 행동할 수 있도록 가벼워야 함.
* 예를들어, 사용자가 콘텐츠에 머물고, 콘텐츠를 보기위해 클릭을 하는 액션은 사용자에게 특별한 행동을 요구하지 않는다. 하지만 얼마나 만족을 했는지와 같은 감정을 파악하기 위해서는 별점, 좋아요 등을 클릭하는 추가적인 액션이 요구됨.  <br>
<br>
* (수현's 경험담) 
    * 회사의 로그 데이터를 보다보면, 클릭이나 스크롤 등의 사용함에 따라 자연스레 수집되는 데이터의 수는 방대하게 수집되고 있으나, 댓글이나 좋아요, sns공유 등의 데이터는 1%도 안될정도로 상당히 적은것을 알수있음. <br>
    (물론 화장품이나 쇼핑 사이트처럼 후기가 큰 영향을 미치는 회사의 경우 이 비율은 조금더 많을 수 있음)

4) 어떤 추천 방식을 사용할 것인가<br>
> 콘텐츠 기반의 추천을 할지, 유사 사용자 기반 추천을할지는 어떤 도메인이냐, 어떤 데이터를 보유하고 있는가에 따라서 다름. <br>
일반적으로 고객 데이터가 부족한 경우 콘텐츠 유사성으로 추천하는 경우가 많고, 고객 데이터가 풍부한 경우 유사 사용자 기반을 주로 사용함.
<br>
* (수현's 경험담)
    * 저희 회사의 경우에도 고객을 식별할 수 있는 데이터가 없어서 주로 콘텐츠 기반으로 추천을 함. <br> 
    예를들어, 뉴스 토픽이나 속보성, 사용자 클릭수 등의 데이터로 제공하는 방식. <br>
    만약 고객에 대한 데이터가 있었더라면 '연예' 기사를 주로 소비하는 사람에게는 해당 기사를 더 많이 노출하는 식으로 사용자들 간의 선호 카테고리 기준으로 그룹핑을 해서 추천해볼 수 있음.

### 데이터 수집하기
1. 인구통계학적 데이터 수집
> 아이디, 성별, 생년월일, 이메일 등
>*  넷플리스의 부사장인 토드 옐린은 다음과 같이 말함
    * 연령, 성별, 거주지 데이터 ? 쓰레기통에나 버려라
 * 예를들어, 나와 비슷한 연령대의 친구라고해서 모두의 취향이 동일한 것은 아니기 때문에. 
 * but, 서비스 초기 단계에서는 이런 데이터를 수집하기 어려우므로, 대부분 인구통계학적 데이터가 활용됨.
 <br>
 
2. 로그 데이터 수집 (Time+Action)
> 사용자가 서비스내에서 행하는 행동을 수집하는 것.
콘텐츠 시청 시간, 특정 콘텐츠에 오랫동안 머무는 행위, 콘텐츠를 시청하는 액션, 콘텐츠를 재방문하는 행동 등이 로그데이터에 해당됨. <br>
여기서 중요한 것은 고객의 어떤 행동이 의미있는 행동이고, 데이터화하였을 때 가치가 발생하느냐를 고민하는 것.
> * 예를들어, 유튜브에서 영상을 클릭하지 않더라도 가만히 머물면 영상이 자동으로 재생되는 것을 볼 수 있는데 , 이때동안 사용자가 영상을 얼마나 봤는지 카운팅을 함.<br>
이 기능은 영상의 플레이 시간을 기준으로 사용자가 콘텐츠에 관심이 있는지를 데이터로 파악할 수 있음.
<br>
> * (수현's 경험담)
    * 개인적으로 유튜브에서의 이 영상을 클릭하지 않는 동안에도 재생되는 시간 또한 데이터로 수집한다는 것이 저에게는 의미있게 다가왔습니다. <br>
    회사에서 영상 서비스에서 사용자의 행동분석을 할 때, 영상 클릭수, 재생수, 탭 이동, 탭 활성/비활성 상태 등으로 측정을 하고는 했는데 <br>
    이렇게 재생된 영상에 얼만큼 머무는지에 대한 데이터도 수집하고 있는지 확인해보고 가능하다면 추가로 통계를 쌓는것도 좋은 방법이라는 생각이 들었습니다.
    

3. 고객 반응 데이터 수집 (Action +Emotion)
> 고객이 콘텐츠에 대한 피드백을 제공하는 것. 흔히 알고 있는 좋아요, 싫어요 또는 별점이 해당됨. <br>
그런데 이 또한 점점 세분화 시키는 추세. 찜하기, 공유하기, 제품보기 등으로 분류.
