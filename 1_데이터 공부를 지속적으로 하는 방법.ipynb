### 주제 : 데이터 공부를 지속적으로 하는 방법
https://brunch.co.kr/@joeflow/28

부제 : 일주일 중 5일을 공부할 수 있게 해준 6가지 기술

6가지 기술 중 가장 공감이 갔던 몇가지만 요약했습니다.
* 뭘 할지 고민하는 시간을 줄여라
>     * 머신러닝 분야는 코드를 짜는 능력도 중요하고, 수학, 확률, 통계, 데이터와 알고리즘까지 공부해야하는 범위가 상당히 넓다. 그러니 자신만의 커리큘럼을 만들어라. 고민만 하다가 시간을 다 보내는것 보다 대충이라도 일주일을 위한 플랜을 만들고 이를 따라가라.

* 가끔은 아무것도 하지 말아라
>    * 배우기 자체는 궁극의 스킬이다. 만약 내가 '배우는 것' 자체를 더 잘할 수 잇다면, 나는 어떤 것이든 더 잘할 수 있다. 그래서 나는 '배우는 능력'을 개발시켜야만 한다. 
    * 집중하기와 분산하기. 최고의 배움은 두 가지가 일어날 때 일어난다. 당신이 샤워를 할 때 최고의 깨달음을 얻는 것과 같은 이치이다. 당신의 공부 루틴 중에 아무것도 안 하기(Doing Nothing)를 더 자주해라!
 
* 공부를 놀이처럼 생각하라
>     * 어린아이가 노는 모습을 보면, 무언가에 매료된 듯한 모습을 보인다. 어른의 세계는 가혹하게도 놀이와 일이 철저히 분리되어 있다. 하지만 공부가 매번 일처럼 느껴진다면, 생 지옥일 것이다. 공부하는 것이 하나의 토픽과 다음으로 갈 수 있는 하나의 과정 이라고 생각하면 모든 것을 다음 단계로 가기 위한 게임처럼 연결할 수 있을 것이다.

------------------------------------------------------------------------------------
* 나의 생각
>   * 요즘 저의 상황에 딱 필요한 멘트들이라 굉장히 와닿았던 글이었습니다. 요즘 저도 통계공부를 먼저할지, 머신러닝 이론을 책으로 한권 훑어볼지 두개를 동시에 할지 머릿속이 복잡했는데 이 글을 보면서 지금은 이미 시작한 통계 공부에 집중하고 그다음 공부는 현재 하던 것을 어느정도 마무리 한 후에 시작해야 겠다고 다짐했습니다. 
    * 공부할 것은 많은데, 업무와 병행하다보니 시간이 부족하다고 느껴져서 쫓기듯 항상 마음이 조급했는데, Doing Nothing이 저에게 여유를 안겨주었습니다. 뇌가 쉴 수 없게 계속해서 주입시키면 효율이 떨어질 수 있고 쉽게 지칠 수 있기 때문에 적절하게 균형을 맞춰야 겠다는 생각을 했습니다. 
    
* 관련된 다른 콘텐츠
    * 이 글을 읽고 또 저의 생각을 작성하면서 제가 읽었던 책 중에 '당신의 뇌는 서두르는 법이 없다' 라는 책이 문득 떠올라서 추천드립니다. 불안한 인간의 마음을 뇌 과학적으로 풀어내어 이성적으로 받아들일 수 있게 도움을 주는 책이라고 생각됩니다.
