## 주제 
지식의 시대 가고 '데이터학습'의 시대. 인재교육, '데이터 리터러시'에 초점을


4차산업혁명 시대에는 연륜과 경력보다 방대한 데이터를 이해하고 활용할 줄 아는 능력이 절실하다. <br>
'데이터 리터러시(해독능력)' 와 새로운 것을 빠르게 학습하는 학습민첩성을 기를 수 있는 새로운 인재 교육방식이 필요하다는 뜻이다.

### 1. 21세기는 지식의 시대가 아니라 학습의 시대
과거에는 지식 그 자체가 중요했지만, 이제 그런 시대는 지났다. 기업들이 대학에서 배운 지식이 실무에도움이 되지 않는다고 보는 이유이기도 하다. <br>
> 또한 경력만 믿고 새로운 것을 익히지 않는 사람도 위험해진다. <br>
하나의 직업 안에서도 필요한 전문성이 빠르게 바뀌므로 오늘 많이 안다는 것이 예전처럼 중요하지 않다. <br>
중요한 것은 새로운 것을 학습하는 능력이다.

### 2. 지칠줄 모르는 학습 능력
학습하는 능력을 좌우하는 역량은 '학습민첩성(Learning Agility)' 이다.<br>
"처음 겪는 새로운 상황에서 경험을 통해 배우고 그렇게 배운 것을 성과 창출에 적용하려고 하는 의지와 능력" 으로 정의된다.

#### < 학습민첩성이 뛰어난 사람들이 공통적으로 보인 행동 5가지 >
* 끊임없이 새로운 것에 도전한다.
* 직접적인 피드백을 구한다.
* 처음 경험하는 상황에서도 비판적으로 사고한다.
* 다양한 사람들과 잘 어울리며 일한다.
* 변화하는 상황에서도 자신의 임무를 성공적으로 수행한다.

#### < 학습민첩성을 기르는 방법 >
* 지적 호기심을 유발하라.
    * 기존의 방식과 절차로도 업무가 가능하지만 좀 더 나은 방법이 없는지 스스로에게 물어야 한다. 
    * 브레인스토밍을 할 때에도 이 같은 사고 훈련을 통해 기존 대비 몇 배 더 많은 아이디어를 낼 수 있다. 
* 복잡하고 방대한 정보 속에서 패턴을 찾아내는 것.
    * ex) 쿠팡은 빅데이터 패턴을 보고 물류창고의 상품 배열을 제품군이 아니라, 소비자군으로 나눠야겠다고 판단.
    * 패턴을 찾도록 하는 방법 => 
    강사의 일방적인 강의가 아니라 케이스 스터디, 프로젝트 제안, 경쟁 프레젠테이션 등
* 성찰 훈련
    * 즉, 배운것, 경험한것, 실행한것을 되짚어보고 복기하는 과정
    * ex) '미생' 의 장그래가 직장생활에서의 경험을 집에돌아와서 되돌아보고 바둑의 원리와 비교하며 자신의 행동을 반성하고 조직의 원리를 깨우쳐가는 모습.
    * ex) 지금 우리가 하고있는 '영감 프로젝트' 도 이 성찰훈련에 해당이되겠네요 ㅎㅎ

### 3. 데이터 리터러시(해석능력) 제고
* 자라(Zara)의 사례
    * 기존 패션 업체들이 스타 디자이너와 브랜드 충성도에 의존한 '흥행사업'을 했다면 자라는 스피드, 가성비, 고객 맞춤으로 승부하는 공급망 사업으로 접근한 것. 
* 예전에는 '경력=인사이트' 식이었다면, 지금은 자유자재로 데이터를 모으고, 분석하고, 활용할 수 있다면 얼마든지 비즈니스 가치를 창출할 수 있는 시대이다.

### 4. 데이터 리터러시를 구성하는 하위 역량

#### 4-1) 플랫폼을 만들 수 있는 통찰력 
하나의 제품이 갖는 가치는 한정적이다. 만들어 팔면 그것으로 거래가 끝난다. <br>
플랫폼은 활용과 거래가 반복되면서 가치가 창출된다. 좋은 플랫폼을 만들면 알고리즘을 지배하고, 데이터를 갖게 되고, 부를 좌우할 수 있게 된다.

* 플랫폼적 사고의 반대편에 있는 것이 전문화(specialization)다.
    * 언제 어떤 기술이나 플랫폼이 출현해서 나의 밥그릇을 뺏어갈지 모르는 시대에 한 가지 주특기만으로 승부를 내려고 하는 것은 위험할 수 있다.
    
* 그러므로, 통찰력을 키워야 한다.
    * 온라인 강의를 통해 선행학습을 한 후 오프라인 강의에서 토론과 평가 등을 하는 '플립러닝 방식' 이 대표적. 
        * 그래서, 저는 모두의연구소에서 하는 10월에 오픈되는 '풀잎스쿨'을 이번엔 꼭 하려고합니다(다짐)
